### INSIGHTFUL DATA VISUALISATION 

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the categories and their associated symbols
categories = {
    'Commodity: AGRICULTURE': ['CC=F', 'KC=F', 'ZC=F', 'CT=F', 'LE=F', 'OJ=F', 'ZS=F', 'SB=F', 'ZW=F'],
    'U.S. Treasury Bonds': ['^FVX', '^IRX', '^TNX', '^TYX'],
    'Commodity: PRECIOUS METALS': ['GC=F', 'PA=F', 'PL=F', 'SI=F'],
    'Commodity: ENERGY': ['EH=F', 'HO=F', 'NG=F', 'BZ=F', 'CL=F'],
    'Market Index': ['^DJI', '^IXIC', '^GSPC', '^NYA', '^RUT','^VIX','^GDAXI', '^FTSE', '^HSI'],
    'Commodity: INDUSTRIAL METALS': ['HG=F']
   
}

# Define the time period for analysis
start_date = '1993-01-01'
end_date = '2023-01-01'

# Function to get historical data for a symbol
def get_data(symbol):
    return yf.download(symbol, start=start_date, end=end_date)['Adj Close']

# Function to calculate and plot relative performance
def plot_relative_performance(category, symbols, metric='cumulative'):
    plt.figure(figsize=(12, 8))

    for symbol in symbols:
        data = get_data(symbol)

        # Calculate daily returns
        daily_returns = data.pct_change()

        if metric == 'cumulative':
            # Calculate cumulative return
            relative_performance = (1 + daily_returns).cumprod() - 1
        elif metric == 'percentage_change':
            # Calculate percentage change
            relative_performance = daily_returns * 100

        # Plot relative performance
        plt.plot(relative_performance, label=symbol)

    plt.title(f'Relative Performance - {category}')
    plt.xlabel('Date')
    plt.ylabel('Performance')
    plt.legend()
    plt.show()

# Iterate through categories and plot relative performance
for category, symbols in categories.items():
    plot_relative_performance(category, symbols)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Define agricultural commodities
agricultural_commodities = ['CC=F', 'KC=F', 'ZC=F', 'CT=F', 'LE=F', 'OJ=F', 'ZS=F', 'SB=F', 'ZW=F']

# Define the time period for analysis
start_date = '1993-01-01'
end_date = '2023-01-01'

# Function to get historical data for a symbol
def get_data(symbol):
    return yf.download(symbol, start=start_date, end=end_date)['Adj Close']

# Create a DataFrame to store adjusted close prices
prices_df = pd.DataFrame({commodity: get_data(commodity) for commodity in agricultural_commodities})

# Combine commodities within the same category
agriculture_data = prices_df.mean(axis=1)

# Perform seasonal decomposition
result = seasonal_decompose(agriculture_data, model='multiplicative', period=365)  # Assuming annual seasonality

# Plot the original, trend, seasonal, and residual components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(agriculture_data, label='Original')
plt.legend()
plt.title('Original Time Series')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the commodities
commodities = [
    'CC=F', 'KC=F', 'ZC=F', 'CT=F', 'LE=F', 'OJ=F', 'ZS=F', 'SB=F', 'ZW=F',
    'EH=F', 'HO=F', 'NG=F', 'BZ=F', 'CL=F', 'HG=F', 'GC=F', 'PA=F', 'PL=F', 'SI=F',
]

# Define the time period for analysis
start_date = '1993-01-01'
end_date = '2023-01-01'

# Function to get historical data for a symbol
def get_data(symbol):
    return yf.download(symbol, start=start_date, end=end_date)['Adj Close']

# Create a DataFrame to store adjusted close prices
prices_df = pd.DataFrame({commodity: get_data(commodity) for commodity in commodities})

# Calculate daily returns
daily_returns = prices_df.pct_change().dropna()

# Calculate the correlation matrix
correlation_matrix = daily_returns.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Commodities')
plt.show()
```

```python
# Calculate volatility
volatility = daily_returns.std()

# Plot volatility comparison
plt.figure(figsize=(12, 8))
volatility.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Volatility of Commodities and Market Indices')
plt.xlabel('Symbols')
plt.ylabel('Volatility')
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define symbols for the selected insights
symbols = ['^DJI', '^IXIC', '^GSPC', '^NYA', '^RUT', '^VIX', '^GDAXI', '^FTSE', '^HSI',
           'CC=F', 'KC=F', 'ZC=F', 'CT=F', 'LE=F', 'OJ=F', 'ZS=F', 'SB=F', 'ZW=F',
           'EH=F', 'HO=F', 'NG=F', 'BZ=F', 'CL=F', 'HG=F', 'GC=F', 'PA=F', 'PL=F', 'SI=F',
           '^FVX', '^IRX', '^TNX', '^TYX']

# Function to get historical data for a symbol
def get_data(symbol, start_date='1990-01-01', end_date='2023-01-01'):
    data = yf.download(symbol, start=start_date, end=end_date)
    return data['Adj Close']

# Loop through symbols and display insights
for symbol in symbols:
    data = get_data(symbol)

    # Insight: Basic Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(data, label=symbol)
    plt.title(f'{symbol} - Historical Prices')
    plt.xlabel('Date')
    plt.ylabel('Adjusted Close Price')
    plt.legend()
    plt.show()
```
