### 5. Display a scatter plot of temperature vs. pressure

### 6. Visualize the distribution of humidity

### 7. Visualize wind speed distribution

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Wind Speed_km/h'], bins=20, edgecolor='black')
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```

### 8. The top 10 relationships between wind speed and weather conditions

```python
# Get the top 10 most frequent weather conditions
top_weather_conditions = df['Weather'].value_counts().nlargest(10).index
df_top_conditions = df[df['Weather'].isin(top_weather_conditions)]
 
# Calculate the average wind speed for each weather condition
average_wind_speed = df_top_conditions.groupby('Weather')['Wind Speed_km/h'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=average_wind_speed.values, y=average_wind_speed.index, palette='viridis')
plt.title('Top 10 Relationships Between Wind Speed and Weather Conditions')
plt.xlabel('Average Wind Speed (km/h)')
plt.ylabel('Weather Condition')
plt.show()
```

### 9. Correlation Analysis

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### 10. Weather Condition Distribution:

```python
plt.figure(figsize=(20,16))
weather_counts = df['Weather'].value_counts()
weather_counts.plot(kind='bar', color='skyblue', width=0.8)  # Adjust the width for better visualization
plt.xlabel('Weather Condition')
plt.ylabel('Frequency')
plt.title('Weather Condition Distribution')
plt.xticks(rotation=90)

# Define bins in intervals of 50
plt.yticks(range(0, max(weather_counts) + 1, 50))

plt.show()
```

### 11. Daily Temperature Patterns

### 13. Temperature over time

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Date/Time'], df['Temp_C'], color='green', alpha=0.7, label='Temperature (Â°C)')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Over Time (Bar Chart)')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

### 14. Temperature Trends Month-wise 

```python
# Convert 'Date/Time' column to datetime format
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
 
# Extract month from 'Date/Time'
df['Month'] = df['Date/Time'].dt.month_name()
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Temp_C', data=df, ci=None)
plt.fill_between(x='Month', y1='Temp_C', data=df, alpha=0.2)  # Fill area under the line
plt.title('Temperature Trends Month-wise (Area Chart)')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.show()
```
